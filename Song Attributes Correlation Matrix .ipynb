```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
song_attributes_df = pd.read_csv ("BillboardFromLast20/songAttributes_1999-2019.csv")
song_attributes_df
```

```python
df = pd.DataFrame(song_attributes_df,columns=['Acousticness','Danceability','Energy', 'Instrumentalness', 'Liveness', 'Loudness','Speechiness','Tempo','Valence'])

corrMatrix = df.corr()
sn.heatmap(corrMatrix, annot=True)
plt.show()
```
