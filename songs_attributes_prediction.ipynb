```python
import numpy as np
import pandas as pd
```

```python
import tensorflow
tensorflow.keras.__version__
```

```python
song_attributes_df = pd.read_csv ("BillboardFromLast20/songAttributes_1999-2019.csv")
song_attributes_df
```

```python
songAttributes=song_attributes_df.drop(columns=["Unnamed: 0","Album","Mode", "Name"])
songAttributes
```

```python
songAttributes_binary_encoded = pd.get_dummies(songAttributes, columns=["Artist", "TimeSignature","Explicit"])
songAttributes_binary_encoded.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
```

```python
songAttributes_binary_encoded.columns
```

```python
X = songAttributes_binary_encoded.drop(["Popularity","Artist_t.A.T.u.", "TimeSignature_1"], axis=1)
y = songAttributes_binary_encoded["Popularity"]
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=1)
X_train.head()
```

```python
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(units=100, activation='relu', input_dim=1004))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=1, activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss='mean_squared_error',
              metrics=['accuracy'])
```

```python
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=60,
    shuffle=True,
    verbose=2
)
```
