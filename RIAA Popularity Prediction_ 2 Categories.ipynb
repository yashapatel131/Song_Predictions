```python
import numpy as np
import pandas as pd
```

```python
import tensorflow
tensorflow.keras.__version__
```

```python
riaaSingle = pd.read_csv('BillboardFromLast20_Merged/merged_riaa_single_2Vlaues.csv')
riaaSingle
```

```python
riaaSingle1=riaaSingle.drop(columns=["X","Name_x", "RiaaStatus","Unique","Label","Unnamed: 0","Album","Artist_y","Mode", "Name_y"])
riaaSingle1
```

```python
riaaSingle_binary_encoded = pd.get_dummies(riaaSingle1, columns=["Artist", "TimeSignature","Explicit"])
riaaSingle_binary_encoded.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
```

```python
X = riaaSingle_binary_encoded.drop(["Popularity","Popularity Calculated", "Artist_Young Dro", "TimeSignature_3","Explicit_False"], axis=1)
y = riaaSingle_binary_encoded["Popularity Calculated"]
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=1)
X_train.head()
```

```python
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(units=100, activation='relu', input_dim=318))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=2, activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=60,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
encoded_predictions = model.predict_classes(X_test_scaled)
prediction_labels = label_encoder.inverse_transform(encoded_predictions)
```

```python
print(f"Predicted classes: {prediction_labels}")
print(f"Actual Labels: {list(y_test)}")
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score 
```

```python
results = confusion_matrix(y_test, prediction_labels) 
print ('Confusion Matrix :')
print(results) 
print ('Accuracy Score :',accuracy_score(y_test, prediction_labels))
print ('Report : ')
print (classification_report(y_test, prediction_labels) )
```
